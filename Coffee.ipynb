```python
import pandas as pd
import numpy as np
dataFrame = pd.DataFrame({'District': ['Kollam ', ' Kottayam ', 'Kochi', 'Wayanad'],
                           
                          'Farm Name': ['GHFarm','ABFarm','CDFarm','EDFarm'],
                           
                          'Aroma': [7.00,7.93,7.12,7.63],
                           
                          'Flavor': [8.50,8.00,8.23,8.45]})                                  
filtered_values = np.where((dataFrame['Aroma']<8) & (dataFrame['Flavor']<=9) & (dataFrame['District'].str.startswith('K')))
print(filtered_values)
display(dataFrame.loc[filtered_values])
```
